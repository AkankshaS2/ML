```python
import numpy as np
import pandas as pd
import math
```

```python
data={'X1':[1,2,3,6],'X2':[1,2,3,6],'Class':['A','B','A','B']}
```

```python
p=input("Enter the point")
p=p.split(',')
a=int(p[0])
b=int(p[1])
```

```python
def distance(a,b):
    l=[]
    # x1,y1
    for i in range (4):
        x1=(data['X1'][i]-a)**2
        y1=(data['X2'][i]-b)**2
        l.append(math.sqrt(x1+y1))
    
    return l;

        
```

```python
l=distance(a,b)
print(l)
k=3
# case
d=[]
used=[]
min=10000
index=79
for j in range(k):
    for i in range(4):
        if l[i]<min and i not in used:
            min=l[i]
            index=i
    used.append(index)
    d.append(min)
    min=10000
print(d,used)    
```

```python
kclass={'A':0,'B':0}
for i in used:
    kclass[data['Class'][i]]+=1;
print('A' if kclass['A']>kclass['B'] else 'B')
```
